```python
data["product"].mode()[0]
{"type:""string"}
data['product']= data['product'].fillna(data["product"].mode()[0])
data["brand"].mode()[0]
{"type":"strings"}
data['brand']= data['brand'].fillna(data["brand"].mode()[0])
data["sale_price"].mean()
np.float64(334.6483912301717)
data['sale_price']= data['sale_price'].fillna(data["sale_price"].mean())
data["rating"].mean()
np.float64(3.943295)
data['rating']= data['rating'].fillna(data["rating"].mean())
data["description"].mode()[0]
{"type:""string"}
data['description']= data['description'].fillna(data["description"].mode()[0])
data["discount"].mean()
np.float64(47.469267850012706)
data['discount']= data['discount'].fillna(data["discount"].mean())

```

```python
import seaborn as sns
```

```python
sns.boxplot(data=data[["index","sale_price","market_price","rating"]])
```

```python
q1=data["sale_price"].quantile(0.25)
q3=data["sale_price"].quantile(0.75)
iqr=q3-q1
iqr
np.float64(264.0)
lower_fence=max(q1-1.5*iqr,0)
upper_fence=q3+1.5*iqr
lower_fence
```

```python
upper_fence
```

```python
outliers=data[(data["sale_price"] < lower_fence) | (data["sale_price"] > upper_fence)]
print("Outliers_count:", len(outliers))
```

```python
data["sale_price"] = np.where(data["sale_price"] <= lower_fence, lower_fence, data["sale_price"])
data["sale_price"] = np.where(data["sale_price"] > upper_fence, upper_fence, data["sale_price"])

sns.boxplot(x=data["sale_price"], color="pink")
plt.title("After capping of outliers")
plt.show()
```

```python
outliers_after_caping = data[(data["sale_price"] < lower_fence) &(data["sale_price"] > upper_fence)]
print("outliers left after caping in sale_price",outliers_after_caping.shape[0])
```
